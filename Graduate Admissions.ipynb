## Split the data into train and test and fit model and test the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X=pd.DataFrame(df,columns=['GRE Score','TOEFL Score','CGPA'])
```

```python
y=pd.DataFrame(df,columns=["Chance of Admit"])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

## Predict the value using model

```python
predition=model.predict(X_test)
```

## Scatter plot for true value vs prediction

```python
plt.scatter(y_test,predition)
plt.title("True values vs Prediction")
plt.xlabel("True Values")
plt.ylabel("Prediction")
plt.show()
```

## r_square score

```python
r2_score(y_test,predition)
```
